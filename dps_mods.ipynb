```python
base_mod_multipliers = {
    "composite": 1.0,
    "overclocked": 1/0.9,
    "dynamic": 1.04,
    "forceful": 1.08,
}
```

```python
nf_mod_mult = {
    "composite": 1.1,
    "overclocked": 1/(1-0.197),
    "dynamic": 1.088,
    "forceful": 1.176,
}
print("EXALTED")
for mod in base_mod_multipliers:
    dps_factor = 1.0
    for mod2 in base_mod_multipliers:
        if mod2 == mod:
            dps_factor *= nf_mod_mult[mod]
        else:
            dps_factor *= base_mod_multipliers[mod2]
    print(mod, f"{dps_factor:.4f}")
```
