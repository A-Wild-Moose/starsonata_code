```python
def calc_factories(n_extractors, extr_rate_per_day, tick_rate, exe_level, ic_use):
    """
    Calculate number of factories required.

    Parameters
    ----------
    n_extractors : int
        Number of equipable extractors.
    extr_rate_per_day : int
        Extraction rate per day
    tick_rate : float
        Tick rate of the item in minutes. 30 seconds would be 0.5
    exe_level : int
        Extraction expert level
    ic_use : int
        Commods used per 1 ic.
    """
    rate_per_min = extr_rate_per_day / (24 * 60) * n_extractors
    rate_per_min *= (1 + 0.22 * exe_level)

    ic_factories = rate_per_min * tick_rate / ic_use

    print(f"IC Factories: {ic_factories:.1f} Rate per minute: {rate_per_min}")
```

```python
# METALS
calc_factories(106, 17280, 48/60, 30, 800)
# calc_factories(53, 34560, 48/60, 30, 800)  # undead metal
calc_factories(115, 34560, 48/60, 30, 800)  # metal fusion (t16)
```
