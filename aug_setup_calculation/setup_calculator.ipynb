```python
import numpy as np
import polars as pl
```

```python
augs = pl.read_csv("parsed_augmenters.csv").drop(
    "Tech"
).with_columns(
    pl.exclude("Name").cast(pl.Float64)
)
```

```python
augs.filter(
    pl.col("Name").is_in([
        "Ult. Obliteration Augmenter",
        "Nightfury's Patience"
    ])
)
```

```python
args1 = (
    pl.when(pl.col(i) <= 0).then(-1 / (pl.col(i) + 1) + 1).otherwise(pl.col(i)).alias(i) for i in augs.columns[1:]
)
# res[res <= 0] = -1 - 1 / (res[res <= 0] - 1)
args2 = (
    pl.when(pl.col(i) <= 0).then(-1 - 1 / (pl.col(i) - 1)).otherwise(pl.col(i)).alias(i) for i in augs.columns[1:]
)

augs.lazy().filter(
    pl.col("Name").is_in([
        "The Emperor's Augmenter",
        "Ult. Obliteration Augmenter",
        "Qa'ik Vazuk Akk'oj"
    ])
).with_columns(
    *args1
).select(
    pl.exclude("Name").sum()
).with_columns(
    *args2
).with_columns(
    dps=(100 * (1 + pl.col("Damage")) * (1 + pl.col("Rate of Fire"))) * (1 + ((0.05 + pl.col("Critical Hit Chance")) * (0.5 * (1 + pl.col("Critical Hit Strength")))))
).collect()
```

```python
args1 = (
    pl.when(pl.col(i) <= 0).then(-1 / (pl.col(i) + 1) + 1).otherwise(pl.col(i)).alias(i) for i in augs.columns[1:]
)
# res[res <= 0] = -1 - 1 / (res[res <= 0] - 1)
args2 = (
    pl.when(pl.col(i) <= 0).then(-1 - 1 / (pl.col(i) - 1)).otherwise(pl.col(i)).alias(i) for i in augs.columns[1:]
)

augs.lazy().filter(
    pl.col("Name").is_in([
        "Celeste Mark Augmenter",
        "Raurava Bhava",
        "Qa'ik Urk'qii Akk'oj"
    ])
).with_columns(
    *args1
).select(
    pl.exclude("Name").sum()
).with_columns(
    *args2
).with_columns(
    dps=(100 * (1 + pl.col("Damage")) * (1 + pl.col("Rate of Fire"))) * (1 + ((0.05 + pl.col("Critical Hit Chance")) * (0.5 * (1 + pl.col("Critical Hit Strength")))))
).collect()
```
