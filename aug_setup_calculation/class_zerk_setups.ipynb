```python
import numpy as np
import polars as pl
```

```python
augs = pl.scan_csv("augmenter_setups_3.csv").filter(
    pl.all_horizontal(pl.col("Damage", "Rate of Fire", "Shield Max").is_not_null()),
    (pl.col("Electric Tempering") <= 0.0) | (pl.col("Electric Tempering").is_null()),
    ~pl.col("").str.contains("Art of Commanding"),
    ~pl.col("").str.contains("Art of Sharpshooting"),
    ~pl.col("").str.contains("Art of Fast Killing"),
    ~pl.col("").str.contains("Art of Gunnery"),
    ~pl.col("").str.contains("Art of Healing"),
    ~pl.col("").str.contains("Art of Engineering"),
    ~pl.col("").str.contains("Art of Stealth"),
).collect()
```

```python
augs = augs.with_columns(
    dps=(100 * (1 + pl.col("Damage")) * (1 + pl.col("Rate of Fire"))) * (1 + ((0.05 + pl.col("Critical Hit Chance")) * (0.5 * (1 + pl.col("Critical Hit Strength")))))
)
```

```python
with pl.Config(fmt_str_lengths=1000, tbl_width_chars=1000):
    display(augs.filter(
        # augs[''].str.contains("Art of Destruction Augmenter, Ultimate") & augs[''].str.contains("Zarkara")
        pl.col("").str.contains("Ultimate Art of Destruction Augmenter', 'Ultimate Art of Destruction Augmenter', 'Zarkara")
    ))
```

```python
a1 = "'Selenite Augmenter'"
a2 = '''"Operative's Cosmic Symmetry Augmenter"'''
a3 = '''"Qa'ik Urk'qii Akk'oj"'''

with pl.Config(fmt_str_lengths=1000, tbl_width_chars=1000):
    display(augs.filter(
        # augs[''].str.contains("Art of Destruction Augmenter, Ultimate") & augs[''].str.contains("Zarkara")
        pl.col("").str.contains(f"{a3}, {a1}, {a2}")
    ))
```

```python
a1 = "'Selenite Augmenter'"
a2 = "'Luminous Resplendent Fluxion Augmenter'"
a3 = "'Ultimate Art of Destruction Augmenter'"

with pl.Config(fmt_str_lengths=1000, tbl_width_chars=1000):
    display(augs.filter(
        # augs[''].str.contains("Art of Destruction Augmenter, Ultimate") & augs[''].str.contains("Zarkara")
        pl.col("").str.contains(f"{a1}, {a2}, {a3}") | \
        pl.col("").str.contains(f"{a2}, {a1}, {a3}") | \
        pl.col("").str.contains(f"{a2}, {a3}, {a1}") | \
        pl.col("").str.contains(f"{a3}, {a2}, {a1}") | \
        pl.col("").str.contains(f"{a1}, {a3}, {a2}") | \
        pl.col("").str.contains(f"{a3}, {a1}, {a2}")
    ))
```

```python
tmp = augs.filter(
    pl.col("dps") >= 650,
    pl.col("Resistance to Damage") >= 0.5,
    pl.col("Shield Max") >= 1.8,
    pl.col("Range") > 0,
    pl.col("Electric Tempering") < 0.0,
    pl.col("Energy Max") > 0.5,
    # pl.col("Energy Charge") > 0.5
).drop(
    "Docking Speed",
    "Hostility",
    "Inertial Dampening",
    "Multifiring",
    "Projectile Velocity",
    "Radar",
    "Tractor Power",
    "Tractor Range",
    "Visibility",
    "Weapon Hold",
    "Weapons Slots"
)

with pl.Config(fmt_str_lengths=1000, tbl_width_chars=1000, tbl_rows=50):
    display(tmp.sort("dps", descending=True))
```
