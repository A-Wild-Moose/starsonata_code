```python
import re
import itertools

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
with open("NF_tracker/raw/rolls_shield.log", 'r') as f:
    _data = f.readlines()

tmp = re.split("(?<=[a-z])([A-Z])", _data[0])

data = [tmp[0]] + ["".join(i) for i in itertools.batched(tmp[1:], 2)]
data = [i.split("|") for i in data]
```

```python
gain = []
loss = []
for j, new in enumerate(data[1:], start=1):
    old = data[j-1]
    
    g = set(new) - set(old)
    l = set(old) - set(new)
    
    if len(g)==1 and len(l) == 1:
        if 'Superconducting' not in list(g)[0]:
            gain.extend(g)
            loss.extend(l)

df = pd.DataFrame({"original": loss, "new": gain})
```

```python
swap_counts = df.value_counts(['original', 'new']).reset_index().sort_values(['original', 'new'])
swap_counts
```
